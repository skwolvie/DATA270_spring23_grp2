```python
df.drop(columns=['start_station_name', 'start_lat',
       'start_lng', 'start_neighborhood', 'nts_rco_T', ], inplace=True)
df['date']= df['date'].astype('datetime64[ns]')
df['start_station_id']= df['start_station_id'].astype('str')
df['start_station_id']= df['start_borough']+'_'+df['start_station_id']
df
```

```python
h= df[['start_station_id', 'start_borough', 'city']].drop_duplicates()
h
```

```python
hierarchy = {}
for index, row in h.iterrows():
    city = 'total'
    borough = row['start_borough']
    station_id = row['start_station_id']
    
    if city not in hierarchy:
        hierarchy[city] = {}
    if borough not in hierarchy[city]:
        hierarchy[city][borough] = []
    hierarchy[city][borough].append(station_id)

hier = {}
for level1, level2 in hierarchy.items():
    if isinstance(level2, dict):
        hier[level1] = list(level2.keys())
        for level3, level4 in level2.items():
            if isinstance(level4, list):
                hier[level3] = level4
```

```python
d= df[['date', 'start_station_id', 'start_borough', 'city', 'ride_count_outflow']]
```

```python
pivot_data = d.pivot_table(index='date', columns='start_station_id', values='ride_count_outflow')
pivot_data.reset_index(inplace=True)

grouped_data = d.groupby(['date', 'start_borough']).agg({'ride_count_outflow': 'sum'}).reset_index()
g_pivot_data= grouped_data.pivot(index='date', columns='start_borough', values='ride_count_outflow')
g_pivot_data.reset_index(inplace=True)

grouped_data = d.groupby(['date', 'city']).agg({'ride_count_outflow': 'sum'}).reset_index()
g2_pivot_data= grouped_data.pivot(index='date', columns='city', values='ride_count_outflow')
g2_pivot_data.reset_index(inplace=True)
g2_pivot_data.columns=['date', 'total']
```

```python
dff= df[['date', 'dayofweek', 'month', 'year', 'day_condition_le', 'temp_mean', 'precip',
       'visibility_mean', 'uvindex_mean', 'nyc_gasprice_avg',
       'traffic_stdS_PC1', 'traffic_stdS_PC2']].drop_duplicates()

dff.reset_index(inplace=True, drop=True)
pivot_data.reset_index(inplace=True, drop=True)
g_pivot_data.reset_index(inplace=True, drop=True)
g2_pivot_data.reset_index(inplace=True, drop=True)

dff['date']= dff['date'].astype('datetime64[ns]')
pivot_data['date']= pivot_data['date'].astype('datetime64[ns]')
g_pivot_data['date']= g_pivot_data['date'].astype('datetime64[ns]')
g2_pivot_data['date']= g2_pivot_data['date'].astype('datetime64[ns]')

dff.sort_values(by='date', inplace=True)
pivot_data.sort_values(by='date', inplace=True)
g_pivot_data.sort_values(by='date', inplace=True)
g2_pivot_data.sort_values(by='date', inplace=True)


dff.reset_index(inplace=True, drop=True)
pivot_data.reset_index(inplace=True, drop=True)
g_pivot_data.reset_index(inplace=True, drop=True)
g2_pivot_data.reset_index(inplace=True, drop=True)

df2= pd.concat([dff, pivot_data.drop(columns=['date']), g_pivot_data.drop(columns=['date']), g2_pivot_data.drop(columns=['date'])], axis=1)
df2
```

```python
dfm= df2.drop(columns=['dayofweek',
       'month', 'year', 'day_condition_le', 'temp_mean', 'precip',
       'visibility_mean', 'uvindex_mean', 'nyc_gasprice_avg',
       'traffic_stdS_PC1', 'traffic_stdS_PC2']).set_index('date')
```
