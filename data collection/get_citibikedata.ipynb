```python
from io import BytesIO
from zipfile import ZipFile
import urllib.request
import re
from joblib import Parallel, delayed
from tqdm import tqdm

def process_zip_file(url):
    if (url.split('/')[-1][:-4] or url.split('/')[-1][:-4]+'.csv') in os.listdir():
        pass
    else:
        try:
            data = []
            with ZipFile(BytesIO(urllib.request.urlopen(url).read())) as my_zip_file:
                for contained_file in my_zip_file.namelist():
                    for line in my_zip_file.open(contained_file).readlines():
                        s = str(line,'unicode_escape')
                        s = re.sub(r"\n", "", s)
                        s = re.sub(r"\"", "", s)
                        line_s = s.split(",")
                        data.append(line_s)
                                
            df = pd.DataFrame(data)
            col_name = df.iloc[0].astype(str)
            df.columns = col_name
            df = df.drop([0])
            if url.split('/')[-1][:-4].endswith('.csv'):
                df.to_csv(url.split('/')[-1][:-4], index=False)
            else:
                df.to_csv(url.split('/')[-1][:-4]+'.csv', index=False)
        except:
            return url
    
# Set the number of parallel jobs
num_jobs = 10

# # Process the ZIP files in parallel using joblib and tqdm
# results= Parallel(n_jobs=num_jobs)(delayed(process_zip_file)(url) for url in tqdm(links))
# print(results)
```

```python
import geopandas as gpd
from shapely.geometry import Point
import pandas as pd
from tqdm import tqdm
from joblib import Parallel, delayed

nta = gpd.read_file('https://data.cityofnewyork.us/api/geospatial/d3qk-pfyz?method=export&format=Shapefile')
boroughs = gpd.read_file(gpd.datasets.get_path('nybb')).to_crs(epsg=4326)

def get_neighborhood_name(point):
    nta_name = nta[nta.contains(point)].ntaname.values
    return nta_name[0] if nta_name else None

def get_borough_name(point):
    borough_name = boroughs[boroughs.contains(point)]['BoroName'].values
    return borough_name[0] if borough_name else None

# define a function that applies get_neighborhood_name and get_borough_name to each row
def apply_functions(row):
    point = row['start_geometry']
    row['neighborhood'] = get_neighborhood_name(point)
    row['borough'] = get_borough_name(point)
    return row

def create_point(row):
    return Point(row['start station longitude'], row['start station latitude'])
```
